```python
#importing libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("/content/drive/MyDrive/Data/Zomato data .csv")
df
```

```python
df.isnull().sum()
```

```python
#converting the datatype into float

def handlerate(value):
  value=str(value).split("/")
  value=value[0]
  return float(value)

df["rate"]=df["rate"].apply(handlerate)
df["rate"]
```

```python
df.dtypes
```

```python
df.columns
```

```python
# type of restaurant
sns.countplot(x=df["listed_in(type)"])
plt.xlabel("type of restaurant")
```

**conclusion:the majority of resataurant order in dining system**

```python
#visualtion the how many votes has each type of restaurant received from customer

grouped_data=df.groupby(df["listed_in(type)"])["votes"].sum()
result=pd.DataFrame({"votes":grouped_data})
plt.plot(result,color="blue",marker="o",mfc="red",mec="black")
plt.title("votes received by each restaurant")
plt.xlabel("type of restaurant",c="red",size=20)
plt.ylabel("votes",c="red",size=20)
```

**conclusion: majority of votes recevied from customers in dining cateogery**

```python
# visualtion received the majority of ratings

plt.hist(x=df["rate"],bins=5)
plt.title("rating distibution")
plt.xlabel("rating")
plt.ylabel("count")
```

**conclusion: majority of restaurant recived the ratings from 3.5 to 4**

```python
#visualtion the average order by couples
couple_data=df["approx_cost(for two people)"]
sns.countplot(x=couple_data)
```

**conclusion:most of couples oreder by apporximate cost of 300 rupees**

```python
#visualtion the which mode recevied the maximum rating

plt.figure(figsize=(6,6))
sns.boxplot(x=df["online_order"],y=df["rate"],data=df)
```

**conclusion:majority of order received from online mode when compare to offline order**

```python
df.head()
```

```python
#visualtion the which restaurant recevied the offline orders

pivot_table=df.pivot_table(index="listed_in(type)",columns="online_order",aggfunc="size",fill_value=0)
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt="d")
plt.title("restaurant received offline orders")
plt.xlabel("online order")
plt.ylabel("type of restaurant")
```

**conclusion:dining restaurant primarliy accept the offline orders**
